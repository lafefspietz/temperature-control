```python
import pyvisa
import json
import time
import os
import numpy as np
from datetime import datetime
```

```python
rm = pyvisa.ResourceManager()
rm.list_resources()
```

```python
tc = rm.open_resource('GPIB0::15::INSTR') #temperature controller instance
```

```python
tc.query('*IDN?')
```

```python
state = {}
state['A'] = float(tc.query('KRDG? A'))
state['B'] = float(tc.query('KRDG? B'))
state['datetime'] = datetime.today().strftime('%Y-%m-%d %H:%M:%S')
state
```

```python
log = [] # list
time_in_minutes = 240 # 180 minutes is 3 hours, 240 minutes is 4 hours
filename = "cooldown-log-data/cooldown_log_" + datetime.today().strftime('%Y_%m_%d_%H%M') + ".json"
for index in range(time_in_minutes):
    print(index)
    state = {}
    state['A'] = float(tc.query('KRDG? A'))
    state['B'] = float(tc.query('KRDG? B'))
    state['datetime'] = datetime.today().strftime('%Y-%m-%d %H:%M:%S')
    log.append(state)
    file_object = open(filename, "w")
    file_object.write(json.dumps(log))
    file_object.close()
    time.sleep(60)
```
